```python
import qubic
from qubic.selfcal_lib import scatter_plot_FP, TES_Instru2coord, Model_Fringes_Ana
from qubic.instrument import QubicInstrument
import numpy as np
import matplotlib.pyplot as plt

def get_dict(args={}):
    
    '''
    Function for modify the qubic dictionary.
    '''
    ### Get the default dictionary
    dictfilename = 'dicts/pipeline_demo.dict'
    d = qubic.qubicdict.qubicDict()
    d.read_from_file(dictfilename)
    for i in args.keys():
        
        d[str(i)] = args[i]
    
    return d

d = get_dict()
q = QubicInstrument(d)
```

```python
baseline = [60, 63]
Model_Ana = Model_Fringes_Ana(q, baseline, 
                                   theta_source=np.deg2rad(0.), 
                                   phi_source=np.deg2rad(0.), 
                                   nu_source=150e9, fwhm=20., amp=1., frame='ONAFP')

xONAFP, yONAFP, fringes = Model_Ana.get_fringes(times_gaussian=True)
```

```python
nprocess = 4

index_per_process = np.array_split(np.arange(992), nprocess)

num_of_proc_per_index = np.zeros(992)
for proc in range(nprocess):
    for i in index_per_process[proc]:
        num_of_proc_per_index[i] = proc

plt.figure(figsize=(12, 8))
scatter_plot_FP(q, xONAFP, yONAFP, num_of_proc_per_index, frame='ONAFP', fig=None, ax=None,
                    s=None, title=f'{nprocess} processes', unit='', cbar=True, fontsize=14)
plt.show()
```
